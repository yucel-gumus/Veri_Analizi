Korelasyon Analizine Başlayalım

```python

# Kategorik sütunları bulalım
categorical_cols = df.select_dtypes(include=['object']).columns

# Sayısal sütunları bulalım
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Sayısal fakat kategorik sütunları bulalım
numeric_but_categorical = []
for col in numeric_cols:
    if df[col].nunique() <= 10:  # Eğer 10'dan az benzersiz değer varsa, kategorik olarak kabul edelim
        numeric_but_categorical.append(col)

# Kategorik fakat kardinal sütunları bulalım
categorical_but_cardinal = []
for col in categorical_cols:
    if df[col].nunique() > 10:  # Eğer 10'dan fazla benzersiz değer varsa, kardinal olarak kabul edelim
        categorical_but_cardinal.append(col)

# Sonuçları yazdıralım
print("Kategorik Sütunlar:", categorical_cols)
print("Sayısal Sütunlar:", numeric_cols)
print("Sayısal Fakat Kategorik Sütunlar:", numeric_but_categorical)
print("Kategorik Fakat Kardinal Sütunlar:", categorical_but_cardinal)
```

```python
#Burada gelir seviyesi için tekrardan bir kutu grafiği ile dağılımına bakalım
plt.figure(figsize=(12,8))
sns.boxplot(x=df["GDP per capita"])
plt.show(block=True)
```

```python
#GDP per capita sütunu için Q1,Q2,Q3 değerini bulalım
# GDP per capita sütununu küçükten büyüğe sıralayalım
sorted_gdp = df['GDP per capita'].sort_values()

# Toplam satır sayısı
total_rows = len(sorted_gdp)
# Q1 pozisyonunu hesaplayalım
q1_position = int(total_rows * 0.25)
q1_value = sorted_gdp.iloc[q1_position - 1]  # Python'da indeksler 0'dan başladığı için bir azaltma yapmamız lazım

# Medyan (Q2) pozisyonunu hesaplayalım
q2_position = int(total_rows * 0.5)
q2_value = sorted_gdp.iloc[q2_position - 1]

# Q3 pozisyonunu hesaplayalım
q3_position = int(total_rows * 0.75)
q3_value = sorted_gdp.iloc[q3_position - 1]

# Sonuçları yazdıralım
print("GDP per capita sütunu için:")
print("Q1 (yüzde 25) =", q1_value)
print("Q2 (medyan, yüzde 50) =", q2_value)
print("Q3 (yüzde 75) =", q3_value)
```

```python
# gelir seviyesinin outlier olması için alt ve üst limiti bulalım
def outlier_thresholds(dataframe: pd.DataFrame, col_name: str, q1=0.25, q3=0.75):

    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquartile_range = quartile3 - quartile1
    upper_limit = quartile3 + 1.5 * interquartile_range
    lower_limit = quartile1 - 1.5 * interquartile_range
    return lower_limit, upper_limit

lower_limit, upper_limit = outlier_thresholds(df, "GDP per capita")
print("Alt sınır:", lower_limit)
print("Üst sınır:", upper_limit)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Bağımsız ve bağımlı değişkenleri ayıralım
X = df.drop(columns=['Overall rank'])  # Bağımsız değişkenler
y = df[['Score','GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']]  # Bağımlı değişken
df.dropna(inplace=True)
# 'Categorical' türündeki sütunları belirleyalim
categorical_columns = df.select_dtypes(include=['category']).columns

# 'Categorical' türündeki sütunları veri setinden çıkaralım hata veriyor  RFE
df.drop(columns=categorical_columns, inplace=True)

# NaN değerleri ortalama değerlerle dolduralım ki her veride çalışsın
df.fillna(df.mean(), inplace=True)

# RFE modelini oluşturalım
model = LinearRegression()
rfe = RFE(model, n_features_to_select=5)  # Seçilecek öznitelik sayısını belirleyelim

# RFE modelini veri setine uygulayalım
fit = rfe.fit(X, y)

# Seçilen özniteliklerin indekslerini alalım
selected_features_indexes = fit.support_

# Seçilen özniteliklerin isimlerini alalım
selected_features = X.columns[selected_features_indexes]

# Sonuçları kontrol edelim
print("Seçilen öznitelikler:")
print(selected_features)
```

```python

# Seçilen özniteliklerin veri setine atayalım
df_selected = df[selected_features]

# Sonuçları kontrol edelim
print(df_selected.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np


# Veriyi seçilen özniteliklerle sınırlayalım
df_selected = df[selected_features]

# Normalleştirme yapalım
scaler = MinMaxScaler()
df_selected = pd.DataFrame(scaler.fit_transform(df_selected), columns=df_selected.columns)

# Polinom ekleyelim GDP per capita kullanıcaz
df_selected['GDP per capita^2'] = df_selected['GDP per capita'] ** 2

# Etkileşim bulunduralım biraz
df_selected['GDP per capita * Social support'] = df_selected['GDP per capita'] * df_selected['Social support']

# Log dönüşümü yapalım
df_selected['GDP per capita_log'] = np.log(df_selected['GDP per capita'])

# İşlenmiş veri setini kontrol edelim
print(df_selected.head())
```
